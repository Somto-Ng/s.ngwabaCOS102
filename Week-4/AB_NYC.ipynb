```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("C:/Users/steve/OneDrive/Desktop/COMPUTER SCIENCE/Year One/Second Semester/COS 102/s.ngwabaCOS102/AB_NYC_2019.csv")
df
```

```python
df['reviews_per_month'] = df['reviews_per_month'].fillna(df['reviews_per_month'].mean)
```

```python
df = df[df['price'] > 0]
```

```python
df.dropna()
```

```python
plt.figure(figsize=(10,6))
sns.scatterplot(data=df, x = "price", y = "availability_365", alpha = 0.5)
plt.title("Availability against Price")
plt.xlabel("Price")
plt.ylabel("Availability")
```

Insight:
The higher the price, the less available it is.

```python
plt.figure(figsize=(10,6))
sns.histplot(df['price'] , bins = 100)
plt.title("Air BnB prices in NYC")
plt.xlabel("Price")
plt.ylabel("Frequency")
```

Insight:
Prices above 200 dollars are scanty.

```python
data = df[df['price'] < 500]

plt.figure(figsize=(10, 6))
sns.boxplot(x= "neighbourhood_group", y ="price", data=data)
plt.title("Price against Borough")
plt.xlabel("Borough")
plt.ylabel("Price")
```

Insight:
Manhattan has the most expensive pricing, while Bronx has the most consistent pricing.
