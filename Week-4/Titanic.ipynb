```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("C:/Users/steve/OneDrive/Desktop/COMPUTER SCIENCE/Year One/Second Semester/COS 102/s.ngwabaCOS102/titanic.csv")
df
```

```python
df = df[df['Parch'] > 0]
```

```python
df = df.drop_duplicates().dropna()
```

```python
df['Age'] = df['Age'].fillna(df['Age'].median())
```

```python
df
```

```python
plt.figure(figsize=(10,6))
sns.scatterplot(data=df, x = "Age", y = "Fare", alpha = 0.5)
plt.title("Fare against Age")
plt.xlabel("Age")
plt.ylabel("Fare")
```

Insight:
Higher fares and younger ages both significantly improved survival chances, while low-fare adults experienced the highest mortality.

```python
survival_counts = df["Survived"].value_counts()
labels = ["Did Not Survive", "Survived"]
colors = ["purple","green"]
plt.pie(survival_counts, labels=labels, autopct='%1.1f%%', colors=colors)
plt.title('Survival Distribution on the Titanic')
plt.axis('equal')
```

Insight:
Higher fares and younger ages both significantly improved survival chances, while low-fare adults experienced the highest mortality.

```python
numeric_df = df.select_dtypes(include=["number"])

plt.figure(figsize = (10,8))
sns.heatmap(numeric_df.corr(), annot = True, cmap = "YlGnBu")
plt.title("Titanic Features Correlation Heatmap")
plt.show()
```

Insights:
Higher social status and higher fares were the strongest predictors of survival, as shown by the significant correlations between Pclass, Fare, and the survival outcome.
