```python
print("1) Quadratic Equation Roots")
print("2) Cubic Equation Roots")
print("3) Quartic Equation Roots")
choice = input("Select 1, 2, or 3: ")

if choice == "1":
    a = float(input("Enter A: "))
    b = float(input("Enter B: "))
    c = float(input("Enter C: "))
    d = (b**2) - (4*a*c)
    if d >= 0:
        root1 = (-b + (d**0.5)) / (2*a)
        root2 = (-b - (d**0.5)) / (2*a)
        print("Roots:", root1, root2)
    else:
        print("Complex roots")

elif choice == "2":
    a = float(input("Enter A: "))
    b = float(input("Enter B: "))
    c = float(input("Enter C: "))
    d = float(input("Enter D: "))
    
    x = 1.0
    for i in range(100):
        fx = a*x**3 + b*x**2 + c*x + d
        fprime = 3*a*x**2 + 2*b*x + c
        if fprime == 0:
            break
        x = x - (fx / fprime)
    print("The root is:", x)

elif choice == "3":
    a = float(input("Enter A: "))
    b = float(input("Enter B: "))
    c = float(input("Enter C: "))
    d = float(input("Enter D: "))
    e = float(input("Enter E: "))
    
    x = 1.0
    for i in range(100):
        fx = a*x**4 + b*x**3 + c*x**2 + d*x + e
        fprime = 4*a*x**3 + 3*b*x**2 + 2*c*x + d
        if fprime == 0:
            break
        x = x - (fx / fprime)
    print("The root is:", x)
```
