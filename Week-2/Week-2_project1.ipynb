```python
name1 = input("Enter first name: ")
age1 = input("Enter first age: ")

name2 = input("Enter second name: ")
age2 = input("Enter second age: ")

print("Before swapping:")
print(name1, "is", age1)
print(name2, "is", age2)

temp = age1
age1 = age2
age2 = temp

print("After swapping:")
print(name1, "is", age1)
print(name2, "is", age2)
```
